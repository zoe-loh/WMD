```python
#making empty dataframe
scored_df= pd.DataFrame(columns = ["subject", "trial_img", "score", "set_size"])

#picking out only the test trials and eliminating "space" keys
test_df = total_df.loc[total_df["phase"] == "test"]
test_df = test_df.replace("space", 0)
#looping through each subject's trials
for subject in range (1,101):
    curr_df = test_df.loc[test_df["subject_nr"] == subject]
    #looping through each test image
    for idx, row in curr_df.iterrows():
        curr_img = row.image
        correct_df = answer_df.loc[answer_df["Img"] == curr_img]
        #converting float to integer
        sub_key = int(row.key)
        if sub_key == correct_df.Answer.values[0]:
            score = 1
        else:
            score = 0
        setSize = row.setSize
        scored_df = scored_df.append(pd.DataFrame({'subject':[subject], 'trial_img': [curr_img], "score": [score], "set_size": [setSize]}))
scored_df
```
